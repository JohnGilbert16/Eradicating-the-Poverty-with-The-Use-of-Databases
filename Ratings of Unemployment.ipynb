```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
%matplotlib inline 
from matplotlib import pyplot as plt
plt.style.use("bmh")
```

```python
df = pd.read_csv(r'C:\Users\clarissa\OneDrive\Documents\Python Scripts\Family_size - Sheet1.csv',delimiter=',')
df.head()
```

```python

b=df.Incidence_M
d=df.Incidence_X


#increase size
plt.figure(figsize=(10,10))
plt.plot(b, color ='red', label ='Insidence 2006')
plt.plot(d, color ='blue', label ='Insidence 2009')

plt.legend()
plt.title("Family Size", fontweight ="bold")
```

```python
e=df.Distribution_Z
f=df.Distribution_R
G=df.Distribution_V

plt.plot(e, color ='brown', label ='Distribution 2003')
plt.plot(f, color ='brown', label ='Distribution 2006')
plt.plot(G, color ='yellow', label ='Distribution 2009')

plt.legend()
plt.title("CDistribution Family Size", fontweight ="bold")
```
