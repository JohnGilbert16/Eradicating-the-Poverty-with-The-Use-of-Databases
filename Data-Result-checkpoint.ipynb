```python
import pandas as pd
import numpy as nd
import matplotlib
%matplotlib inline 
from matplotlib import pyplot as plt
plt.style.use("bmh")
```

```python
df = pd.read_csv(r'C:\Users\clarissa\OneDrive\Documents\Python Scripts\consolidated_philippines_poverty_data.csv')
df.head()
```

```python
x =df.agr_wage_farm_workers_allgender_2015	
y= df.agr_wage_farm_workers_male_2015	
z= df.agr_wage_farm_workers_female_2015
plt.plot(x, c ='orange', label ='All Gender')
plt.plot(y, c= 'green', label='Male Worker')
plt.plot(z, c ='yellow', label = 'Female Worker')
plt.legend()
plt.title("Gender Comparison of workers", fontweight ="bold")
```
